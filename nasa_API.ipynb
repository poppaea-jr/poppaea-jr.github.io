```python
# Using Python to access data from an API
```

```python
import requests
import pandas as pd
import json
```

```python
# requesting data from my API URL
```

```python
response = requests.get("https://power.larc.nasa.gov/api/temporal/daily/point?parameters=WS10M&community=SB&longitude=51.45904975&latitude=-2.603436197922635&start=20210110&end=20220110&format=JSON")
```

```python
# checking that the fetch was sucessful
```

```python
print(response.status_code)
```

```python
print(response.json())
```

```python
data = response.json()
data
```

```python
from pandas import json_normalize
```

```python
# using json normalise to load my json data into a flat table
```

```python
df = json_normalize(data)
```

```python
df
```

```python
df = df.T
```

```python
df
```

```python
# dropping meta information for tidy data
```

```python
df = df.iloc[4: , :]
```

```python
df
```

```python
df = df.iloc[:-12]
```

```python
df
```

```python
df.info()
```

```python
df = df.reset_index()
```

```python
df
```

```python
df = df.rename(columns={0: "wind"})
```

```python
df
```

```python
df = df.rename(columns={"index": "date"})
```

```python
df
```

```python
# using a for loop to iterate over the date column, to isolate the date information
```

```python
for x in range(0,len(df)):
  df['date'][x] = df['date'][x][27:]
```

```python
df
```

```python
import pandas as pd
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
df
```

```python
df.to_csv("senate_wind.csv", index = False)
```
